Change one-hot encoded data (Race, Religion and Location) back to data grouped into 1 feature only so that LIME and SHAP local only returns it as 1 feature in the explanation and summary plots. (Ex: Location instead of Location_Penang)

```python
# using inverse_transform method from scikit learn

import pandas as pd

df = pd.read_csv('b_all.csv')

def inverse_one_hot(df, prefix):
    cols = [c for c in df.columns if c.startswith(prefix + "_")]
    return df[cols].idxmax(axis=1).str.replace(prefix + "_", "", regex=False)

# Recover original columns
df["Race"] = inverse_one_hot(df, "Race")
df["Religion"] = inverse_one_hot(df, "Religion")
df["Location"] = inverse_one_hot(df, "Location")

# Optionally drop one-hot columns
df = df.drop(columns=[c for c in df.columns if
                      c.startswith("Race_") or
                      c.startswith("Religion_") or
                      c.startswith("Location_")])

print(df[["Race", "Religion", "Location"]].head())
```
