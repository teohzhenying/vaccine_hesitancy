<h2>KNN</h2>

KNN normalized

```python
def knn_normalized():
    
    max_accuracy = 0
    max_x = 1


    for x in range(1, 31):

        clf = KNeighborsClassifier(n_neighbors=x)

        scores = cross_val_score(clf, x_train_normalized, y_train_normalized,
                                 cv=5, scoring='accuracy')
        score_mean = scores.mean()
        if score_mean > max_accuracy:
            max_accuracy = score_mean
            max_x = x
        
    print("Best accuracy is ", max_accuracy, " at p = ", max_x)

knn_normalized()
```

```python
# Train the model using the entire training set with the best k
knn_best = KNeighborsClassifier(n_neighbors=10)
knn_best.fit(x_train_normalized, y_train_normalized)
```

```python
# Predict the target for the test dataset
test_predict = knn_best.predict(x_test_normalized)

test_predict
```

```python
# Compute the model accuracy on the test set: How often is the classifier correct?
accuracy = metrics.accuracy_score(y_test_normalized, test_predict)
print("Accuracy: ", accuracy)
```

<h2>Decision tree</h2>

Decision tree normalized

```python
from sklearn.tree import DecisionTreeClassifier


def dt(x_train, y_train):

    # Initialize lists to store depth values and validation scores
    D = []
    validation = []
    scores = {}
    
    # The for loop will test depth values from 3 to 22
    for d in range(3, 23):
        clf = DecisionTreeClassifier(criterion='entropy', max_depth=d, random_state=0)
    
        # Perform cross-validation and compute the mean accuracy
        cv_scores = cross_val_score(clf, x_train, y_train, cv=5, scoring='accuracy')
        validation_score = cv_scores.mean()
    
        # Store depth value and corresponding validation score
        D.append(d)
        validation.append(validation_score)
        scores[d] = validation_score
    
    # Find the best depth
    best_depth = max(scores, key=scores.get)
    print(f"The best depth is {best_depth} with a validation score of {scores[best_depth]}")
    
    return best_depth


best_depth = dt(x_train_normalized, y_train_normalized)
```

```python
# Train the model using the entire training set with the best depth
dtree_best = DecisionTreeClassifier(criterion='entropy', max_depth=best_depth, random_state=0)
dtree_best.fit(x_train_normalized, y_train_normalized)
```

```python
# Predict the target for the test dataset
test_predict = dtree_best.predict(x_test_normalized)

test_predict
```

```python
# Compute the model accuracy on the test set: How often is the classifier correct?
accuracy = metrics.accuracy_score(y_test_normalized, test_predict)
print("Accuracy: ", accuracy)
```

<h2>MLP - 1 layer</h2>

MLP 1 layer normalized

```python
# Import necessary libraries
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier

# Function to perform cross-validation and find the best hidden layer size
def find_best_hidden_layer_size(hidden_layer_sizes, x_train, y_train):
    best_score = 0
    best_size = None
    for size in hidden_layer_sizes:
        mlp = MLPClassifier(hidden_layer_sizes=size, max_iter=3000, random_state=42)
        scores = cross_val_score(mlp, x_train, y_train, cv=5, scoring='accuracy')
        mean_score = scores.mean()
        print(f'Hidden layer size {size}: Mean cross-validation accuracy = {mean_score}')
        if mean_score > best_score:
            best_score = mean_score
            best_size = size
    return best_size, best_score

# Define hidden layer sizes to test
hidden_layer_sizes_1 = [(h,) for h in [5] + list(range(10, 101, 5))]
hidden_layer_sizes_2 = [(5, h) for h in [5] + list(range(10, 101, 5))]

# Find the best hidden layer size for 1 hidden layer
best_size_1, best_score_1 = find_best_hidden_layer_size(hidden_layer_sizes_1, x_train_normalized, y_train_normalized)
print(f'Best hidden layer size for 1 layer: {best_size_1} with score {best_score_1}')
```

<h2>SVM (linear)</h2>

SVM normalized

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter range for C
C_range = [0.1,1,10,100]
degree = [2,3,4]

param_grid_linear = {'C': C_range}

# Function to perform Grid Search and evaluate the model for linear kernel
def evaluate_linear_kernel(x_train, y_train, x_test, y_test):
    print("Evaluating linear kernel")
    grid_search = GridSearchCV(SVC(kernel='linear'), param_grid_linear, cv=5, scoring='accuracy')
    grid_result = grid_search.fit(x_train, y_train)
    print(f"Best parameters for linear kernel: {grid_result.best_params_}")

    # Train the model with the best parameters
    best_model = grid_result.best_estimator_
    best_model.fit(x_train, y_train)

    # Predict on the test set
    test_predict = best_model.predict(x_test)

    # Compute and print the model accuracy on the test set
    accuracy = metrics.accuracy_score(y_test, test_predict)
    print(f"Accuracy (Test) for linear kernel: {accuracy}")

    # Print confusion matrix and classification report
    print(confusion_matrix(y_test, test_predict))
    print(classification_report(y_test, test_predict))
    print("\n" + "="*60 + "\n")

#print the best evaluate linear kernel    
evaluate_linear_kernel(x_train_normalized, y_train_normalized, x_test_normalized, y_test_normalized)
```
