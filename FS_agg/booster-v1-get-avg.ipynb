```python
import pandas as pd
```

```python
df_normalized_before_processing = pd.read_csv('normalized_vaccine_data_processed.csv')
df_not_normalized_before_processing = pd.read_csv('not_normalized_vaccine_data_processed.csv')
```

```python
A = ["S.VaccineImportantHealth",
     "S.VaccineEffectiveSevere",
     "S.VaccineImportantPublic",
     "S.VaccineBeneficialMOH",
     "S.VaccineSafe",
     "S.MOHInfoTrusted",
     "S.VaccineProtectInfection",
     "S.AgreeMOHRecommendation",
     "S.WorriedSideEffects",
     "S.PreventNewVariant"]

B = ["P.ProtectFlu",
     "P.ProtectSevereComp",
     "P.EffectivenessVsRisk",
     "P.ClinicallyTested",
     "P.LongTermSideEffects",
     "P.TrustReportsMOH",
     "P.ChipBelief",
     "P.MandatoryBelief",
     "P.HalalDoubt",
     "P.AlternativeMedicineBelief",
     "P.HighRiskJob",
     "P.WorkDemand"]

C = ["A.BoosterProtectSevere",
     "A.BoosterProtectFamily",
     "A.BoosterPreventSpread",
     "A.ReturnDailyActivities",
     "A.RecommendingBooster",
     "A.MaintainAntibody",
     "A.NoSeriousSideEffects"]

D = ["B.RiskWithoutBooster",
     "B.SevereCompWithoutBooster",
     "B.SpreadFamilyWithoutBooster",
     "B.TransmitPatientsWithoutBooster",
     "B.SkepticalBooster",
     "B.BoosterSafe",
     "B.BoosterMoreSideEffects",
     "B.PreferNaturalImmunity",
     "B.EliminateStigma",
     "B.PublicRiskWithoutBooster"]

E = ["C.BoosterNotNeeded",
     "C.BoosterHarm",
     "C.EvidenceInsufficient",
     "C.TrustSocialMedia",
     "C.SideEffectsNoImpact"]
```

```python
# get the mean of all the A, B, C, D and E columns

ABCDE_columns = A + B + C + D + E

# 1) Make a copy so you don't mutate the original
df_work = df_normalized_before_processing.copy()

# 2) Create aggregated columns directly
df_work["A_mean"] = df_work[A].mean(axis=1)
df_work["B_mean"] = df_work[B].mean(axis=1)
df_work["C_mean"] = df_work[C].mean(axis=1)
df_work["D_mean"] = df_work[D].mean(axis=1)
df_work["E_mean"] = df_work[E].mean(axis=1)

# 3) Save the target, drop raw ABCDE items + target, then add target back
y = df_work["AgreeSubsequentBooster"]  # keep it aside

df_normalized = df_work.drop(columns=ABCDE_columns + ["AgreeSubsequentBooster"], errors="ignore")
df_normalized["AgreeSubsequentBooster"] = y
```

```python
# do the same for not normalized

# 1) Make a copy so you don't mutate the original
df_work_ii = df_not_normalized_before_processing.copy()

# 2) Create aggregated columns directly
df_work_ii["A_mean"] = df_work_ii[A].mean(axis=1)
df_work_ii["B_mean"] = df_work_ii[B].mean(axis=1)
df_work_ii["C_mean"] = df_work_ii[C].mean(axis=1)
df_work_ii["D_mean"] = df_work_ii[D].mean(axis=1)
df_work_ii["E_mean"] = df_work_ii[E].mean(axis=1)

# 3) Save the target, drop raw ABCDE items + target, then add target back
y = df_work_ii["AgreeSubsequentBooster"]  # keep it aside

df_not_normalized = df_work_ii.drop(columns=ABCDE_columns + ["AgreeSubsequentBooster"], errors="ignore")
df_not_normalized["AgreeSubsequentBooster"] = y
```

```python
# import to csv

df_normalized.to_csv('booster_normalized.csv', index=False)
```
