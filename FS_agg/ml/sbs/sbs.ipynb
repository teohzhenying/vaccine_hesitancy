<h1>SFS</h1>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
import warnings
import numpy as np

# Suppress overflow warning
warnings.filterwarnings("ignore", category=RuntimeWarning)
np.seterr(over='ignore')
```

```python
df_normalized = pd.read_csv('booster_normalized.csv')
target_normalized = df_normalized['AgreeSubsequentBooster']

features_normalized = df_normalized.drop('AgreeSubsequentBooster', axis=1)

x_train_normalized, x_test_normalized, y_train_normalized, y_test_normalized = train_test_split(features_normalized,
                                                                                                target_normalized,
                                                                                                test_size=0.2,
                                                                                                random_state=0)
```

__KNN__

KNN normalized

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=10)

sfs_knn = SFS(knn,
              k_features=(1, len(x_train_normalized.columns)),
              forward=False,
              floating=False,
              scoring='accuracy',
              cv=5)

sfs_knn = sfs_knn.fit(x_train_normalized, y_train_normalized)

selected_indices_knn = sfs_knn.k_feature_idx_
selected_features_knn = [x_train_normalized.columns[i] for i in selected_indices_knn]

print('\nSequential Forward Selection for KNN (Normalized):')
print('Selected feature indices:', selected_indices_knn)
print('Selected feature names:', selected_features_knn)
print('Total number of selected features:', len(selected_features_knn))
print('CV Score:', sfs_knn.k_score_)

x_train_sfs_knn = x_train_normalized[selected_features_knn]
knn.fit(x_train_sfs_knn, y_train_normalized)

x_test_sfs_knn = x_test_normalized[selected_features_knn]
y_pred_knn = knn.predict(x_test_sfs_knn)

final_accuracy_knn = accuracy_score(y_test_normalized, y_pred_knn)
print('Final accuracy for KNN on test set (Normalized):', final_accuracy_knn)
```

__Decision tree__

Decision tree normalized

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

decision_tree = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=0)

sfs_dt = SFS(decision_tree,
             k_features=(1, len(x_train_normalized.columns)),
             forward=False,
             floating=False,
             scoring='accuracy',
             cv=5)

sfs_dt = sfs_dt.fit(x_train_normalized, y_train_normalized)

selected_indices_dt = sfs_dt.k_feature_idx_
selected_features_dt = [x_train_normalized.columns[i] for i in selected_indices_dt]

print('\nSequential Forward Selection for Decision Tree (Normalized):')
print('Selected feature indices:', selected_indices_dt)
print('Selected feature names:', selected_features_dt)
print('Total number of selected features:', len(selected_features_dt))
print('CV Score:', sfs_dt.k_score_)

x_train_sfs_dt = x_train_normalized[selected_features_dt]
decision_tree.fit(x_train_sfs_dt, y_train_normalized)

x_test_sfs_dt = x_test_normalized[selected_features_dt]
y_pred_dt = decision_tree.predict(x_test_sfs_dt)

final_accuracy_dt = accuracy_score(y_test_normalized, y_pred_dt)
print('Final accuracy for Decision Tree on test set (Normalized):', final_accuracy_dt)
```

mnb normalized
