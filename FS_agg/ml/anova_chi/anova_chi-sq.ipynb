<h1>Booster - ANOVA + Chi2</h1>

```python
import pandas as pd
from sklearn import metrics, svm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.svm import SVC
```

```python
import warnings
import numpy as np

# Suppress overflow warning
warnings.filterwarnings("ignore", category=RuntimeWarning)
np.seterr(over='ignore')
```

```python
df_normalized = pd.read_csv('anova_chi_normalized.csv')

target_normalized = df_normalized['AgreeSubsequentBooster']



features_normalized = df_normalized.drop('AgreeSubsequentBooster', axis=1)

x_train_normalized, x_test_normalized, y_train_normalized, y_test_normalized = train_test_split(features_normalized,
                                                                                                target_normalized,
                                                                                                test_size=0.2,
                                                                                                random_state=0)
```

__KNN__

KNN normalized

```python
# Train the model using the entire training set with the best k
knn_best = KNeighborsClassifier(n_neighbors=10)
knn_best.fit(x_train_normalized, y_train_normalized)

# Predict the target for the test dataset
test_predict = knn_best.predict(x_test_normalized)

accuracy = metrics.accuracy_score(y_test_normalized, test_predict)
print("Final Accuracy: ", accuracy)
```

__Decision tree__

Decision tree normalized

```python
from sklearn.tree import DecisionTreeClassifier

# Train the model using the entire training set with the best depth
dtree = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=0)
dtree.fit(x_train_normalized, y_train_normalized)

# Predict the target for the test dataset
test_predict = dtree.predict(x_test_normalized)

# Compute the model accuracy on the test set: How often is the classifier correct?
accuracy = metrics.accuracy_score(y_test_normalized, test_predict)
print("Accuracy: ", accuracy)
```

__GNB__

gnb normalized

```python
current_model = GaussianNB()

current_model.fit(x_train_normalized, y_train_normalized)
y_pred = current_model.predict(x_test_normalized)

acc = metrics.accuracy_score(y_test_normalized, y_pred)
print("Accuracy (Test):", acc)
```

__MNB__

mnb normalized

```python
mnb = MultinomialNB()

scores = cross_val_score(mnb, x_train_normalized, y_train_normalized, cv = 10, scoring="accuracy")
print("Accuracy (Validation) =", scores.mean())
mnb.fit(x_train_normalized, y_train_normalized)
test_predict = mnb.predict(x_test_normalized)
print("Accuracy (Test): ", metrics.accuracy_score(y_test_normalized, test_predict))
```

__MLP 1 layer__

MLP 1 layer normalized

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(30,), max_iter=3000, random_state=42)
mlp.fit(x_train_normalized, y_train_normalized)
test_predict = mlp.predict(x_test_normalized)
print("Accuracy (Test): ", metrics.accuracy_score(y_test_normalized, test_predict))
```

__MLP 2 layer__

MLP 2 layer normalized

```python
mlp = MLPClassifier(hidden_layer_sizes=(5, 40), max_iter=3000, random_state=42)
mlp.fit(x_train_normalized, y_train_normalized)
test_predict = mlp.predict(x_test_normalized)
print("Accuracy (Test): ", metrics.accuracy_score(y_test_normalized, test_predict))
```

__SVM (linear)__

SVM normalized

```python
model = svm.LinearSVC(C=0.1, random_state=42) # 'C' is the regularization parameter

# 3. Train the model
model.fit(x_train_normalized, y_train_normalized)

# 4. Make predictions
predictions = model.predict(x_test_normalized)

# 5. Evaluate the model
accuracy = accuracy_score(y_test_normalized, predictions)
print(f"Model Accuracy: {accuracy:.2f}")
```

__SVM (poly)__

SVM poly normalized

```python
# 1. Create the polynomial SVM model
model = svm.SVC(
    kernel='poly',
    C=1,
    degree=3,
    gamma='auto'
)

# 2. Train the model
model.fit(x_train_normalized, y_train_normalized)

# 3. Make predictions
predictions = model.predict(x_test_normalized)

# 4. Evaluate the model
accuracy = accuracy_score(y_test_normalized, predictions)
print(f"Model Accuracy: {accuracy:.2f}")
```

__SVM (rbf)__

SVM rbf normalized

```python
model = svm.SVC(
    kernel='rbf',
    C=10,
    gamma='scale'
)

model.fit(x_train_normalized, y_train_normalized)

predictions = model.predict(x_test_normalized)

accuracy = accuracy_score(y_test_normalized, predictions)
print(f"Model Accuracy: {accuracy:.2f}")
```

__SVM sigmoid__

SVM sigmoid normalized

```python
model = svm.SVC(
    kernel='sigmoid',
    C=10,
    gamma='auto'
)

model.fit(x_train_normalized, y_train_normalized)

predictions = model.predict(x_test_normalized)

accuracy = accuracy_score(y_test_normalized, predictions)
print(f"Model Accuracy: {accuracy:.2f}")
```
