```python
# test set accuracy values

acc_vals = {"SHAP_Model_1": 0.825, "SHAP_Model_2": 0.825, "SHAP_Model_3": 0.9, "SHAP_Model_4": 0.65, "SHAP_Model_5": 0.825, "SHAP_Model_6": 0.65}
```

```python
import pandas as pd
```

```python
df = pd.read_csv("SHAP_values.csv")


final_df = pd.DataFrame()
final_df["Feature_Name"] = df["Feature"]

# calculation for phase 5 to get final global features importance
final_df["Final_Global_Feature_Importance"] = (
        df["SHAP_Model_1"] * acc_vals["SHAP_Model_1"] +
        df["SHAP_Model_2"] * acc_vals["SHAP_Model_2"] +
        df["SHAP_Model_3"] * acc_vals["SHAP_Model_3"] +
        df["SHAP_Model_4"] * acc_vals["SHAP_Model_4"] +
        df["SHAP_Model_5"] * acc_vals["SHAP_Model_5"] +
        df["SHAP_Model_6"] * acc_vals["SHAP_Model_6"]
)
```

```python
# sort values
final_df = final_df.sort_values(by="Final_Global_Feature_Importance", ascending=False)
```

```python
final_df.to_csv(f"final_global_results.csv", index=False)
```
