```python
import pandas as pd

# f(x) weight (lime and shap)
fx_weight = [[0.613379430077077, 0.765], [0.525230167469948, 0.413], [0.659672137469874, 0.1],
             [0.43269312005310906, 0.284], [0.337179390277098, 0.419], [0.344687068849189, 0.1]]

# scores data
scores_data = ["./data/all_dt.csv", "./data/all_svm_rbf.csv", "./data/anova_chi_gnb.csv", "./data/base_agg_svm_poly_bagging.csv", "./data/anova_chi_agg_boosting.csv", "./data/anova_chi_agg_gnb.csv"]

# for printing
print_statements = ["Base model - ALL Features (Decision Tree)", "Base model - ALL Features (SVM rbf)", "Anova and Chi Square (GNB)", "Base model - Aggregated (SVM Poly, Bagging)", "Anova and Chi Square - Aggregated (Boosting)", "Anova and Chi Square - Aggregated (GNB)"]
```

```python
output_names = [
    "all_dt.csv",
    "all_svm_rbf.csv",
    "anova_chi_gnb.csv",
    "base_agg_svm_poly_bagging.csv",
    "anova_chi_agg_boosting.csv",
    "anova_chi_agg_gnb.csv"
]

for index in range(6):
    curr_csv = pd.read_csv(scores_data[index])

    print("Final feature importance for", print_statements[index], ":")

    # compute fusion score
    curr_csv["Final_Score"] = (
            fx_weight[index][0] * curr_csv["LIME_Score"].abs() +
            fx_weight[index][1] * curr_csv["SHAP_Score"].abs()
    )

    # sort descending
    curr_csv = curr_csv.sort_values(by="Final_Score", ascending=False)

    # print
    print(curr_csv[["Feature", "Final_Score"]])
    print("\n")

    # save to csv
    curr_csv.to_csv(f"./final_data/{output_names[index]}", index=False)
```
