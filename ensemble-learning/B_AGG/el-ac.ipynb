```python
# import statements
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("anova_chi_normalized.csv")
target = df["AgreeSubsequentBooster"]
features = df.drop("AgreeSubsequentBooster",axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

__Random Forest__

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics

best_n_estimator = 1
best_scores = 0

for n in range(1, 20):
    # Random Forest Bagging
    model_bagging_rf = BaggingClassifier(estimator=RandomForestClassifier(max_features="log2", bootstrap=False), n_estimators=n, random_state=7)
    
    # Cross-validation
    print("Model: Random Forest")
    scores_rf = cross_val_score(model_bagging_rf, x_train, y_train, cv=5, scoring='accuracy')
    print("Accuracy scores for each fold: ", scores_rf)
    print("Mean Accuracy (Validation):", scores_rf.mean())
    print("-" * 50)
    
    # find the best n_estimator
    if scores_rf.mean() > best_scores:
        best_scores = scores_rf.mean()
        best_n_estimator = n
    
print("Best n estimator is: ", best_n_estimator, " with score ", best_scores)

# Test set evaluation
model_bagging_rf = BaggingClassifier(estimator=RandomForestClassifier(max_features="log2", bootstrap=False), n_estimators=best_n_estimator, random_state=7)
model_bagging_rf.fit(x_train, y_train)
test_predict_rf = model_bagging_rf.predict(x_test)
print("Random Forest Bagging - Accuracy (Test):", metrics.accuracy_score(y_test, test_predict_rf))
print("=" * 50)
```

__Bagging - GNB__

```python
# Import necessary libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics

# Neural Network with one hidden layer of 5 neurons
model_bagging_dt = BaggingClassifier(
    estimator=GaussianNB(),
    n_estimators=10,
    random_state=7
)

# Cross-validation
print("Model: GNB")
scores_nn = cross_val_score(model_bagging_dt, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold: ", scores_nn)
print("Mean Accuracy (Validation):", scores_nn.mean())
print("-" * 50)

# Test set evaluation
model_bagging_dt.fit(x_train, y_train)
test_predict_nn = model_bagging_dt.predict(x_test)
print("GNB Bagging - Accuracy (Test):", metrics.accuracy_score(y_test, test_predict_nn))
print("=" * 50)
```

__Bagging - SVM poly__

```python
# Import necessary libraries
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics

# SVC 
model_bagging_poly = BaggingClassifier(
    estimator=SVC(kernel="poly", C= 1, degree= 3, gamma= "auto"),
    n_estimators=5,
    random_state=7
)

# Cross-validation
print("Model: SVC poly")
scores_nn = cross_val_score(model_bagging_poly, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold: ", scores_nn)
print("Mean Accuracy (Validation):", scores_nn.mean())
print("-" * 50)

# Test set evaluation
model_bagging_poly.fit(x_train, y_train)
test_predict_nn = model_bagging_poly.predict(x_test)
print("SVC poly Bagging - Accuracy (Test):", metrics.accuracy_score(y_test, test_predict_nn))
print("=" * 50)
```

__Bagging - 2 hidden__

```python
# Import necessary libraries
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics

# Neural Network with one hidden layer of 5 neurons
model_bagging_nn = BaggingClassifier(
    estimator=MLPClassifier(hidden_layer_sizes=(5,40), max_iter=10000, random_state=7),
    n_estimators=10,
    random_state=7
)

# Cross-validation
print("Model: Neural Network (2 hidden layer)")
scores_nn = cross_val_score(model_bagging_nn, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold: ", scores_nn)
print("Mean Accuracy (Validation):", scores_nn.mean())
print("-" * 50)

# Test set evaluation
model_bagging_nn.fit(x_train, y_train)
test_predict_nn = model_bagging_nn.predict(x_test)
print("Neural Network Bagging - Accuracy (Test):", metrics.accuracy_score(y_test, test_predict_nn))
print("=" * 50)
```

__Bagging - SVM sigmoid__

```python
# Import necessary libraries
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics

# SVC 
model_bagging_sigmoid = BaggingClassifier(
    estimator=SVC(kernel="sigmoid", C= 10, gamma= "auto"),
    n_estimators=5,
    random_state=7
)

# Cross-validation
print("Model: SVC sigmoid")
scores_nn = cross_val_score(model_bagging_sigmoid, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold: ", scores_nn)
print("Mean Accuracy (Validation):", scores_nn.mean())
print("-" * 50)

# Test set evaluation
model_bagging_sigmoid.fit(x_train, y_train)
test_predict_nn = model_bagging_sigmoid.predict(x_test)
print("SVC sigmoid Bagging - Accuracy (Test):", metrics.accuracy_score(y_test, test_predict_nn))
print("=" * 50)
```

__Boosting__

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Define the Gradient Boosting classifier
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Cross-validation
cv_scores = cross_val_score(gbc, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold (CV): ", cv_scores)
print("Mean Accuracy (Validation): {:.4f}".format(cv_scores.mean()))
print("-" * 50)

# Fit the model on the full training data
gbc.fit(x_train, y_train)

# Evaluate the model on the test data
y_test_pred = gbc.predict(x_test)
accuracy = accuracy_score(y_test, y_test_pred)

# Print evaluation metrics
print(f'Accuracy (Test): {accuracy:.4f}')
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_test_pred))
print('Classification Report:')
print(classification_report(y_test, y_test_pred))
print("=" * 50)
```

__XGBoost__

```python
import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np

# Define the XGBoost classifier
xgb_clf = xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Cross-validation
cv_scores = cross_val_score(xgb_clf, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold (CV): ", cv_scores)
print("Mean Accuracy (Validation): {:.4f}".format(cv_scores.mean()))
print("-" * 50)

# Fit the model on the full training data
xgb_clf.fit(x_train, y_train)

# Evaluate the model on the test data
y_test_pred = xgb_clf.predict(x_test)
accuracy = accuracy_score(y_test, y_test_pred)

# Print evaluation metrics
print(f'Accuracy (Test): {accuracy:.4f}')
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_test_pred))
print('Classification Report:')
print(classification_report(y_test, y_test_pred))
print("=" * 50)
```
