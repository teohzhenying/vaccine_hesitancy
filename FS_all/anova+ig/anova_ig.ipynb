<h1>Booster - ANOVA + IG</h1>

```python
import pandas as pd
from sklearn import metrics, svm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.svm import SVC
```

```python
import warnings
import numpy as np

# Suppress overflow warning
warnings.filterwarnings("ignore", category=RuntimeWarning)
np.seterr(over='ignore')
```

```python
df_normalized = pd.read_csv('anova_cin_normalized.csv')
df_not_normalized = pd.read_csv('anova_cin_not_normalized.csv')
target_normalized = df_normalized['AgreeSubsequentBooster']

target_not_normalized = df_not_normalized['AgreeSubsequentBooster']

features_normalized = df_normalized.drop('AgreeSubsequentBooster', axis=1)

features_not_normalized = df_not_normalized.drop('AgreeSubsequentBooster', axis=1)
x_train_normalized, x_test_normalized, y_train_normalized, y_test_normalized = train_test_split(features_normalized,
                                                                                                target_normalized,
                                                                                                test_size=0.2,
                                                                                                random_state=0)

x_train_not_normalized, x_test_not_normalized, y_train_not_normalized, y_test_not_normalized = train_test_split(
    features_not_normalized, target_not_normalized, test_size=0.2, random_state=0)
```

__KNN__

KNN normalized

```python
# Train the model using the entire training set with the best k
knn_best = KNeighborsClassifier(n_neighbors=10)
knn_best.fit(x_train_normalized, y_train_normalized)

# Predict the target for the test dataset
test_predict = knn_best.predict(x_test_normalized)

accuracy = metrics.accuracy_score(y_test_normalized, test_predict)
print("Final Accuracy: ", accuracy)
```

KNN not normalized

```python
# Train the model using the entire training set with the best k
knn_best = KNeighborsClassifier(n_neighbors=4)
knn_best.fit(x_train_not_normalized, y_train_not_normalized)

# Predict the target for the test dataset
test_predict = knn_best.predict(x_test_not_normalized)

accuracy = metrics.accuracy_score(y_test_not_normalized, test_predict)
print("Final Accuracy: ", accuracy)
```

__Decision tree__

Decision tree normalized

```python
from sklearn.tree import DecisionTreeClassifier

# Train the model using the entire training set with the best depth
dtree = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=0)
dtree.fit(x_train_normalized, y_train_normalized)

# Predict the target for the test dataset
test_predict = dtree.predict(x_test_normalized)

# Compute the model accuracy on the test set: How often is the classifier correct?
accuracy = metrics.accuracy_score(y_test_normalized, test_predict)
print("Accuracy: ", accuracy)
```

```python
# Train the model using the entire training set with the best depth
dtree = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=0)
dtree.fit(x_train_not_normalized, y_train_not_normalized)

# Predict the target for the test dataset
test_predict = dtree.predict(x_test_not_normalized)

# Compute the model accuracy on the test set: How often is the classifier correct?
accuracy = metrics.accuracy_score(y_test_not_normalized, test_predict)
print("Accuracy: ", accuracy)
```
