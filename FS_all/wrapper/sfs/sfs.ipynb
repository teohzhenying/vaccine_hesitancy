```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
import warnings
import numpy as np

# Suppress overflow warning
warnings.filterwarnings("ignore", category=RuntimeWarning)
np.seterr(over='ignore')
```

```python
df_normalized = pd.read_csv('normalized_vaccine_data_processed.csv')
df_not_normalized = pd.read_csv('not_normalized_vaccine_data_processed.csv')
target_normalized = df_normalized['AgreeSubsequentBooster']

target_not_normalized = df_not_normalized['AgreeSubsequentBooster']

features_normalized = df_normalized.drop('AgreeSubsequentBooster', axis=1)

features_not_normalized = df_not_normalized.drop('AgreeSubsequentBooster', axis=1)
x_train_normalized, x_test_normalized, y_train_normalized, y_test_normalized = train_test_split(features_normalized,
                                                                                                target_normalized,
                                                                                                test_size=0.2,
                                                                                                random_state=0)

x_train_not_normalized, x_test_not_normalized, y_train_not_normalized, y_test_not_normalized = train_test_split(
    features_not_normalized, target_not_normalized, test_size=0.2, random_state=0)
```

__MLB 1 layer__

MLB 1 layer normalized

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

mlp_norm = MLPClassifier(hidden_layer_sizes=(50,), random_state=0, max_iter=10000)

sfs_mlp_norm = SFS(mlp_norm,
                   k_features=(1, len(x_train_normalized.columns)),
                   forward=True,
                   floating=False,
                   scoring='accuracy',
                   cv=5)

sfs_mlp_norm = sfs_mlp_norm.fit(x_train_normalized, y_train_normalized)
selected_indices_mlp_norm = sfs_mlp_norm.k_feature_idx_
selected_features_mlp_norm = [x_train_normalized.columns[i] for i in selected_indices_mlp_norm]

x_train_mlp_sfs_norm = x_train_normalized[selected_features_mlp_norm]
mlp_norm.fit(x_train_mlp_sfs_norm, y_train_normalized)

x_test_mlp_sfs_norm = x_test_normalized[selected_features_mlp_norm]
y_pred_mlp_norm = mlp_norm.predict(x_test_mlp_sfs_norm)
final_accuracy_mlp_norm = accuracy_score(y_test_normalized, y_pred_mlp_norm)

print('Final accuracy for MLP 1 layer (Normalized):', final_accuracy_mlp_norm)
```

__SVM (linear)__

SVM normalized

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

linear_svm = SVC(kernel='linear', C=0.1, random_state=42)

sfs_linear_svm = SFS(linear_svm,
                     k_features=(1, len(x_train_normalized.columns)),
                     forward=True,
                     floating=False,
                     scoring='accuracy',
                     cv=5)

sfs_linear_svm = sfs_linear_svm.fit(x_train_normalized, y_train_normalized)
selected_indices_linear = sfs_linear_svm.k_feature_idx_
selected_features_linear = [x_train_normalized.columns[i] for i in selected_indices_linear]

print('\nSequential Forward Selection for Linear SVM (Normalized):')
print('Selected feature indices:', selected_indices_linear)
print('Selected feature names:', selected_features_linear)
print('Total number of selected features:', len(selected_features_linear))
print('CV Score:', sfs_linear_svm.k_score_)

x_train_sfs_linear = x_train_normalized[selected_features_linear]
linear_svm.fit(x_train_sfs_linear, y_train_normalized)

x_test_sfs_linear = x_test_normalized[selected_features_linear]
y_pred_linear = linear_svm.predict(x_test_sfs_linear)

final_accuracy_linear = accuracy_score(y_test_normalized, y_pred_linear)
print('Final accuracy on test set for Linear SVM (Normalized):', final_accuracy_linear)
```
