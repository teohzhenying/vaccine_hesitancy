<h1>SBS</h1>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
import warnings
import numpy as np

# Suppress overflow warning
warnings.filterwarnings("ignore", category=RuntimeWarning)
np.seterr(over='ignore')
```

```python
df_normalized = pd.read_csv('normalized_vaccine_data_processed.csv')
df_not_normalized = pd.read_csv('not_normalized_vaccine_data_processed.csv')
target_normalized = df_normalized['AgreeSubsequentBooster']

target_not_normalized = df_not_normalized['AgreeSubsequentBooster']

features_normalized = df_normalized.drop('AgreeSubsequentBooster', axis=1)

features_not_normalized = df_not_normalized.drop('AgreeSubsequentBooster', axis=1)
x_train_normalized, x_test_normalized, y_train_normalized, y_test_normalized = train_test_split(features_normalized,
                                                                                                target_normalized,
                                                                                                test_size=0.2,
                                                                                                random_state=0)

x_train_not_normalized, x_test_not_normalized, y_train_not_normalized, y_test_not_normalized = train_test_split(
    features_not_normalized, target_not_normalized, test_size=0.2, random_state=0)
```

__KNN__

KNN normalized

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=10)

# Apply SBS to select the best features
sbs_knn = SFS(knn,
              k_features=(1, len(x_train_normalized.columns)),
              forward=False,     # BACKWARD selection
              floating=False,
              scoring='accuracy',
              cv=5)

# Fit the SBS model
sbs_knn = sbs_knn.fit(x_train_normalized, y_train_normalized)

# Get the selected features
selected_indices_knn = sbs_knn.k_feature_idx_
selected_features_knn = [x_train_normalized.columns[i] for i in selected_indices_knn]

print('\nSequential Backward Selection for KNN:')
print('Selected feature indices:', selected_indices_knn)
print('Selected feature names:', selected_features_knn)
print('Total number of selected features:', len(selected_features_knn))
print('CV Score:', sbs_knn.k_score_)

# Use the selected features for training
x_train_sbs_knn = x_train_normalized[selected_features_knn]
knn.fit(x_train_sbs_knn, y_train_normalized)

# Test the model on the test set
x_test_sbs_knn = x_test_normalized[selected_features_knn]
y_pred_knn = knn.predict(x_test_sbs_knn)

# Calculate accuracy
final_accuracy_knn = accuracy_score(y_test_normalized, y_pred_knn)

print('Final accuracy for KNN on test set:', final_accuracy_knn)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Initialize the Polynomial SVM classifier
poly_svm_norm = SVC(kernel='poly', C=0.1, degree=2, gamma='scale')

# Apply SBS (Backward Selection) for Polynomial SVM (Normalized)
sbs_poly_svm_norm = SFS(poly_svm_norm,
                        k_features=(1, len(x_train_normalized.columns)),
                        forward=False,        # Sequential Backward Selection
                        floating=False,
                        scoring='accuracy',
                        cv=5)

# Fit the SBS to the Polynomial SVM
sbs_poly_svm_norm = sbs_poly_svm_norm.fit(x_train_normalized, y_train_normalized)
selected_indices_poly_sbs_norm = sbs_poly_svm_norm.k_feature_idx_
selected_features_poly_sbs_norm = [x_train_normalized.columns[i] for i in selected_indices_poly_sbs_norm]

print('\nSequential Backward Selection for Polynomial SVM (Normalized):')
print('Selected feature indices:', selected_indices_poly_sbs_norm)
print('Selected feature names:', selected_features_poly_sbs_norm)
print('Total number of selected features:', len(selected_features_poly_sbs_norm))
print('CV Score:', sbs_poly_svm_norm.k_score_)

# Train and evaluate the Polynomial SVM
x_train_poly_sbs_norm = x_train_normalized[selected_features_poly_sbs_norm]
poly_svm_norm.fit(x_train_poly_sbs_norm, y_train_normalized)
x_test_poly_sbs_norm = x_test_normalized[selected_features_poly_sbs_norm]
y_pred_poly_sbs_norm = poly_svm_norm.predict(x_test_poly_sbs_norm)
final_accuracy_poly_sbs_norm = accuracy_score(y_test_normalized, y_pred_poly_sbs_norm)

print('Final accuracy on test set for Polynomial SVM (Normalized):', final_accuracy_poly_sbs_norm)
```
